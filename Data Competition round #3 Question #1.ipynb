## visualize taking Highest Qualification as label

```python
u_df = df.copy()
g_df = df.copy()
p_df = df.copy()

u_df = u_df.loc[u_df['Highest Qualification'] != 'Undergraduate']
g_df = g_df.loc[g_df['Highest Qualification'] != 'Graduate']
p_df = p_df.loc[p_df['Highest Qualification'] != 'Postgraduate']

u_df = u_df.drop(['Highest Qualification', 'Gender'], axis = 1)
g_df = g_df.drop(['Highest Qualification', 'Gender'], axis = 1)
p_df = p_df.drop(['Highest Qualification', 'Gender'], axis = 1)

col_array = u_df.columns

avg1 = np.zeros(len(col_array))
avg2 = np.zeros(len(col_array))
avg3 = np.zeros(len(col_array))

for i in range(len(col_array)):
    
    avg1[i] = u_df[col_array[i]].mean()
    avg2[i] = g_df[col_array[i]].mean()
    avg3[i] = p_df[col_array[i]].mean()

    
fig2, ax2 = plt.subplots(figsize = (8, 4), dpi = 100)
ax2.plot([i for i in range(len(col_array))], avg1, 'r', label = 'Undergraduate')
ax2.plot([i for i in range(len(col_array))], avg2, 'g', label = 'Graduate')
ax2.plot([i for i in range(len(col_array))], avg3, 'b', label = 'Postgraduate')

ax2.set_ylabel('Average Value')
ax2.set_title('Average Value of Parameters for Male')
ax2.legend()
plt.show()
```

## Analysis

Now we can see that for feature 3, 4, 5 and 6 and 11 also all 3 type of volentiers share the same opinion that yes,  "People have the desire to change  but not sure what to contribute, how and when to contribute" AND "It feels good to get involved in the social activities" AND "People will participate in social activities if their opinions are valued"

But No to, "The cultural limitation doesn't allow the user to participate in social good initiatives" AND "User want s to participate but wants to keep their identity secret" AND "People will participate in social activities if their opinions are valued"

Feature 2 i.e., "No direct benefit/reputation for participating in social activities" has diffent opinions i.e., undergraduates think that this the reason the are not participating in social activities but for Graduate and Postgraduate this is not the case

Also from feature 8 we can see that undergraduates are not scared of politics for participating in social activities while the fear of politics can be clearly seen in graduates and postgraduates. 

Also feature 1 shows that with age you think that you have something to contribute in social activities

and feature 10 shows that with age everyone starts understanding how to deal with personal issues and not to let them stop you from participating in social activities.
